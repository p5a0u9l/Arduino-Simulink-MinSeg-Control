```python
# create symbols
t, x, xdot, x1, x2, u, y = symbols("t, x, \dot{x}, x_1, x_2, u, y")
X, Phi = symbols("X, Phi")
fm1, fm2, t0 = symbols("fm1, fm2, t0")
x0, x0_1, x0_2 = symbols("x0, x0_1, x0_2")
```

```python
# create and display state matrices
A = Matrix([[0, t], [0, 2]])
B = Matrix([[0], [t]])
C = Matrix([0, 1]).transpose()
D = Matrix([0])
A, B, C, D
```

```python
# Compute and display (1)
x1 = Function('x1')(t)
x2 = Function('x2')(t)
xdot = Function('\dot{x}')(t)
x = Matrix([x1, x2])
Eq(xdot, MatAdd(A*x, B*u))
```

```python
# For the ZIR, u(t) = 0
Eq(xdot, MatAdd(A*x, B*u).subs(u, 0))
```

```python
#Solve the system of ODEs, solve for x2 first
dx2 = diff(x2)
eq2 = Eq(dx2, 2*x2)
dsolve(eq2)
```

```python
# Subsitute in the solution for x2 and solve x1
c1, c2, tau = symbols("c1, c2, tau")
x2 = c1*exp(2*t)
Eq(x1, c2 + integrate(tau*x2.subs(t, tau), (tau, 0, t)))
```

```python
# Put x together
x1 = c2 + c1/4 + exp(2*t)*(2*c1*t - c1)/4
Eq(x, Matrix([x1, x2]))
```

### Solution 1.a

```python
# look at initial conditions [0, 1] and [1, 0]
X = Function('X')(t)
fm1 = Matrix([x1, x2]).subs(c1, 1).subs(c2, 0)
fm2 = Matrix([x1, x2]).subs(c1, 0).subs(c2, 1)
X = fm2.row_join(fm1)
X
```

### Solution 1.b

```python
# Find STM (State Transition Matrix) by multiplying FM with it's inverse
X = fm1.row_join(fm2)
Phi = simplify(X*X.inv().subs(t, t0))
Phi
```

### Solution 3.a

#### The state transition matrix is $\mathscr{L}^{-1}\left[ \left(s \mathbf{I}_n - \mathbf{A} \right)^{-1} \right]$ 

```python
sI_A = s*eye(2) - A
inverse_laplace_transform(sI_A.inv(), s, t)
```

### Solution 3.b

```python
Phi = Matrix([[(sin(2*t)+cos(2*t))*exp(-2*t), sin(2*t)*exp(-2*t)], 
             [-2*exp(-2*t)*sin(2*t), sin(2*t)+cos(2*t)*exp(-2*t)]])
Phi
```

```python
# term 1
simplify(C*Phi*x0)
```

```python
# term 2
C*integrate(Phi.subs(t, tau)*B*u, (tau, 0, t))
```

```python
Eq(y, simplify(C*Phi*x0 + C*integrate(Phi.subs(t, tau)*B*u, (tau, 0, t)))[0])
```
