# Solution

## Part a

Let state variable $\mathbf{x}$ be
\begin{equation}
    \mathbf{x} = 
        \begin{vmatrix}
            \alpha \\
            \dot{\alpha} \\
            x \\
            \dot{x} \\
        \end{vmatrix}
        = 
        \begin{vmatrix}
            x_1 \\
            x_2 \\
            x_3 \\
            x_4 \\
        \end{vmatrix}
\end{equation}

### Solve using Python and SymPy package

```python
# Initialize variables
from sympy import *
init_printing()
Ip, mp, L, x1, x3, x2, x4, x, alpha = symbols("I_p, m_p, L, x_1, x_3, x_2, x_4, x, alpha")
g, rw, Iw, mw, Tm, xd, alpha_d = symbols("g, r_w, I_w, m_w, T_m, \dot{x}, \dot{alpha}")
alpha_dd, x_dd = symbols("\ddot{a}, \ddot{x}")
kt, V, R, kb, torque = symbols("k_t, V, R, k_b, torque")
```

#### Solve Eq (1) for $\ddot{\alpha}$ and $\ddot{x}$

```python
# State space variable, x
x = Matrix([[x1], [x2], [x3], [x4]])
x
```

```python
# elements of Eq 1
M1 = Matrix([[-(Ip + mp*L**2), mp*L*cos(alpha)], [mp*L*rw**2*cos(alpha), -(Iw+mw*rw**2+mp*rw**2)]])
v1 = Matrix([[alpha_dd], [x_dd]])
v2 = Matrix([[Tm - mp*L*g*sin(alpha)],[Tm*rw + mp*L*rw**2*alpha**2*cos(alpha)]])
eq1 = Eq(M1*v1, v2)
eq1
```

#### Substituting $\alpha^{eq} = 0$ and letting higher order terms be 0, Eq (1) becomes

```python
eq1 = eq1.subs(alpha, 0)
eq1
```

#### Solve and simplify

```python
s = simplify(solve(eq1, [alpha_dd, x_dd]))
s
```

## Part b

#### The torque equation is given as

```python
torque = kt/R*V + (kt*kb)/(R*rw) + (kt*kb)/R*x2
torque.as_expr()
```

#### where $x_2$ is $\dot{\alpha}$. 

#### Substituting $T_m$ into ${\ddot{\alpha}}$ and ${\ddot{x}}$ and simplifying gives

```python
alpha_dd = simplify(s[alpha_dd].subs(Tm, torque))
x_dd = simplify(s[x_dd].subs(Tm, torque))
(alpha_dd, x_dd)
```

#### The time-derivative of the state-space variable, $\mathbf{\dot{x}}$, is \begin{equation}     \mathbf{\dot{x}} =          \begin{vmatrix}             \dot{\alpha} \\             \ddot{\alpha} \\             \dot{x} \\             \ddot{x} \\         \end{vmatrix}         =          \begin{vmatrix}             x_2 \\             \ddot{\alpha} \\             x_4 \\             \ddot{x} \\         \end{vmatrix} \end{equation}  Substituting ${\ddot{\alpha}}$ and ${\ddot{x}}$ gives

```python
xd = Matrix([x2, alpha_dd, x4, x_dd])
xd
```

## Part c

#### The state matrix $\mathbf{A}$ is the jacobian of $\mathbf{f}$ with respect to $\mathbf{x}$

```python
A = xd.jacobian(x)
A
```

#### The state matrix $\mathbf{B}$ is the jacobian of $\mathbf{f}$ with respect to $V$

```python
B = xd.jacobian([V])
B
```

### Since there is no disturbance, $\mathbf{D}$ is zero  

```python
y, X, C, D = symbols("y, X, C, D")
y = Matrix([[x1], [x3]])
y
```

### Since $\mathbf{y} = [\alpha x]^T = [x_1 x_3]^T$ is the desired output and $\mathbf{y} = \mathbf{C} \mathbf{x}$, 

```python
C = Matrix([[1, 0, 0, 0], [0, 0, 1, 0]])
C
```
